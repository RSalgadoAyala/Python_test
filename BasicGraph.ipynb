```python
from matplotlib import pyplot as plt
```

```python
# data
years=[1950, 1960, 1970,1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
# create a line chart
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title('Nominal GDP')
plt.ylabel('Billions of $')
plt.show()
```

```python
# data
movies = ['Annie Hall', 'Ben-Hur', 'Casablanca', 'Gandhi', 'West Side Story']
num_oscars = [5,11,3,8,10]
```

```python
#plot bars
plt.bar(range(len(movies)), num_oscars)
plt.title('Favorites films')
plt.ylabel('# of Academy Awards')
plt.xticks(range(len(movies)), movies)
plt.show()
```

```python
# histograms
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

#Bucket grades by decile, but put in with the 90s
histogram = Counter(min(grade // 10*10,90) for grade in grades)

plt.bar([x+5 for x in histogram.keys()],histogram.values(),10,edgecolor=(0,0,0))        
plt.axis([-5,105,0,5])                    # first 2 for x, second 2 for y
plt.xticks([10*i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('# of students')
plt.title('Distribution of Exam 1 Grades')
plt.show()
```

```python
mentions = [500, 505]
years = [2017, 2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("No. of times I heard someone say 'data science'")

plt.ticklabel_format(useOffset=False)

plt.axis([2016.5,2018.5,499, 506])
plt.title('Look at the "Huge" increase!')
plt.show()
```

```python
plt.bar(years, mentions, 0.8)  # 0.8 is the thickness of the bars
plt.xticks(years)
plt.ylabel("No. of times I heard someone say 'data science'")

plt.ticklabel_format(useOffset=False)

plt.axis([2016.5,2018.5,0, 550])
plt.title('Look at the not so "Huge" increase!')
plt.show()
```
