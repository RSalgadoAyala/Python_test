```
x=np.linspace(-3,3,100)
y=x**3+2*x**2-x-1

fig=plt.figure() # we create the figure object
ax=fig.add_subplot() # an ax is attached to the figure object where we are going to plot data
ax.plot(x,y,marker='v') # characteristics of the figure

ax.set_xlim(-1,3)
ax.set_ylim()
ax.set_xlabel('x')
ax.set_ylabel('y')
```
